```python
#AQI Calculation and Categorization
#The notebook applies predefined AQI thresholds to pollutant values (such as PM2.5) to categorize air quality into levels like Good, Moderate, and Poor, following environmental standards.
```

```python
sns.heatmap(df.drop(columns=["AQI"]).corr(), annot=True)
plt.title("Air Quality Heat Map")
plt.show()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
def calculate_aqi(pm25):
    if pm25 <= 50:
        return "Good"
    elif pm25 <= 100:
        return "Moderate"
    elif pm25 <= 200:
        return "Poor"
    else:
        return "Very Poor"
```

```python
df["AQI"] = df["pm2_5"].apply(calculate_aqi)
```

```python
df
```

```python
#AQI Assessment Against Standards
#The calculated AQI levels are compared with standard air quality benchmarks to assess whether the air quality is safe or harmful to public health.
```

```python
df["AQI_Status"] = df["AQI"].value_counts()
```
