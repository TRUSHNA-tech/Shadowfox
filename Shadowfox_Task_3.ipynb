```python
plt.plot(df['injuries_no_indication'], df['num_units'], marker='o')
plt.title("Line Plot Title")
plt.xlabel("injuries_no_indication")
plt.ylabel("num_units")
plt.show()
```

```python
plt.scatter(df['num_units'], df['injuries_total'])
plt.show()
```

```python
sns.barplot(
    x=df['crash_day_of_week'],
    y=df['injuries_total']
)
plt.title("Average Injuries by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Average Injuries")
plt.show()
```

```python
avg_injuries = df.groupby('crash_hour')['injuries_total'].mean().reset_index()

sns.lineplot(
    x='crash_hour',
    y='injuries_total',
    data=avg_injuries
)
plt.title("Average Injuries by Crash Hour")
plt.xlabel("Crash Hour")
plt.ylabel("Average Injuries")
plt.show()
```

```python
sns.histplot(
    df['injuries_total'],
    bins=30
)
plt.title("Distribution of Total Injuries")
plt.xlabel("Total Injuries")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(
    x='traffic_control_device',
    y='injuries_total',
    data=df
)
plt.title("Injuries by Traffic Control Device")
plt.xticks(rotation=90)
plt.show()
```

```python
weather_counts = df['weather_condition'].value_counts().head(6)

plt.pie(
    weather_counts,
    labels=weather_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Accidents by Weather Condition")
plt.show()
```

```python
pivot_table = pd.pivot_table(
    df,
    values='injuries_total',
    index='crash_day_of_week',
    columns='crash_hour',
    aggfunc='mean'
)

sns.heatmap(
    pivot_table
)
plt.title("Average Injuries by Day and Hour")
plt.xlabel("Crash Hour")
plt.ylabel("Day of Week")
plt.show()
```
